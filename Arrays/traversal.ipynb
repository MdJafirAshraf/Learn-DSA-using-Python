```python
# Array
arr = [10, 20, 30, 40, 50]
```

```python
# Array Traversal
# Traversal = visiting each element once

arr = [10, 20, 30, 40]

for i in range(len(arr)):
    print(arr[i])

# Time: O(n)
# Space: O(1)
```

```python
# Method 1: Using index
for i in range(len(arr)):
    print(arr[i])
```

```python
# Method 2: Using direct element access
for element in arr:
    print(element)
```

```python
# Method 3: Reverse traversal
for i in range(len(arr)-1, -1, -1):
    print(arr[i])
```

```python
# Example
# 1: Find maximum element - Time: O(n)

arr = [10, 5, 20, 8, 30]

max_element = arr[0]
for i in range((len(arr))):
    if arr[i] > max_element:
        max_element = arr[i]
        
print("Maximum element is:", max_element)
```

```python
# 2: Search an element (Linear Search) - Time: O(n)

arr = [10, 5, 20, 8, 30]
target = 20

for i in range(len(arr)):
    if arr[i] == target:
        print("Element found at index:", i)
        break
```
